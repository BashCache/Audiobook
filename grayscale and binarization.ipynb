```python
import os
original_path="image/train_cleaned/"
augmented_path="image/aug_cleaned/"
images=[]
for im in os.listdir(original_path):  # read image name from folder and append its path into "images" array     
    images.append(os.path.join(original_path,im))
#print(images)
print('\n\n\n')
aug = "image/aug/"
aug_img = []
for im in os.listdir(aug):  # read image name from folder and append its path into "images" array     
    aug_img.append(os.path.join(aug,im))
print(aug_img)
print(os.getcwd())
```

```python
train_clean_images = []
for i in range(len(aug_img)):
    pic = aug_img[i].split('/')
    #print(pic[2])
    img = pic[2].split('-')
    img0 = img[0].split('_')
    #picture = cv2.imread("image/"+img[1])
    #cv2.imwrite(augmented_path)
    aug_no_string = img0[1]
    #print(img0)
    #print(aug_no)
    aug_no = int(aug_no_string)
    #print(img)
    img1 = img[1]
    if aug_no>= 101 and aug_no<1000:
        print('aug_no=',aug_no)
        print(img1,',', aug_no)
    if aug_no>= 101 and aug_no<1000:
        train_clean_images.append(img1)
        
    
print(train_clean_images)
    
```
